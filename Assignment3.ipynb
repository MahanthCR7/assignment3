Responsive design support – works well on mobile devices

Q3.Define accessibility in the context of web development. Discuss why it's essential to create accessible
websites and how it benefits different user group"

Ans:Accessibility in web development means designing websites so that all users, including people with disabilities, can easily access and use them.

It is essential because it ensures equal access to information and services for everyone. It also improves user experience, reaches a wider audience, and may meet legal requirements.

Benefits:

Helps visually impaired users (screen readers)

Supports users with hearing disabilities (captions)

Assists people with motor difficulties (easy navigation)

Improves usability for all users, including beginners and elderly

Q4. List any 3 ways which help us in improving the accessibility of HTML

Q5.Create a web page that highlights the features of HTML5 using semantic tags

```python
<!DOCTYPE html>
<html>
<head>
    <title>HTML5 Features</title>
</head>
<body>

<header>
    <h1>Features of HTML5</h1>
</header>

<section>
    <article>
        <h2>Semantic Elements</h2>
        <p>HTML5 provides tags like header, footer, and article for better structure.</p>
    </article>

    <article>
        <h2>Audio and Video</h2>
        <p>Supports audio and video without external plugins.</p>
    </article>

    <article>
        <h2>Canvas</h2>
        <p>Used to draw graphics and animations.</p>
    </article>
</section>

<footer>
    <p>End of Page</p>
</footer>

</body>
</html>
```
